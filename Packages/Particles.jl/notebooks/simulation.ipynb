```julia
using Revise
using StaticDistributions
using Particles
using StateSpaceModelExamples
using PlotlyJS
```

```julia
ssm = InvariantLinearGaussianProcess_Example_1();
```

```julia
states, observations = rand(ssm, nothing, 1_000);
```

```julia
times = axes(states, 1)
traces = []
for i in eachindex(first(states))
    push!(traces, scatter(; x=times, y=getindex.(states, i), name="X$i", mode="lines"))
end
for i in eachindex(first(observations))
    push!(traces, scatter(; x=times, y=getindex.(observations, i), name="Y$i", mode="lines"))
end
plot([traces...])
```
